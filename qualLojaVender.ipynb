### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
fatu_loja1 = loja["Preço"].sum()
fatu_loja2 = loja2["Preço"].sum()
fatu_loja3 = loja3["Preço"].sum()
fatu_loja4 = loja4["Preço"].sum()

print("Faturamento Loja 1: R$", fatu_loja1)
print("Faturamento Loja 2: R$", fatu_loja2)
print("Faturamento Loja 3: R$", fatu_loja3)
print("Faturamento Loja 4: R$", fatu_loja4)
```

```python
import matplotlib.pyplot as plt

faturamentos = [fatu_loja1, fatu_loja2, fatu_loja3, fatu_loja4]
lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]

plt.figure(figsize=(8, 8))
plt.pie(faturamentos, labels=lojas, autopct='%1.1f%%', startangle=90, colors=plt.cm.Pastel1.colors)

plt.title("Distribuição do Faturamento por Loja", fontsize=16)

plt.show()
```

# 2. Vendas por Categoria


```python
Quantidade1 = loja['Categoria do Produto'].value_counts()
print(Quantidade1)

Quantidade2 = loja2['Categoria do Produto'].value_counts()
print(Quantidade2)

Quantidade3 = loja3['Categoria do Produto'].value_counts()
print(Quantidade3)

Quantidade4 = loja4['Categoria do Produto'].value_counts()
print(Quantidade4)
```

```python
#MELHOR VIZUALIZAÇÃO

df_quantidade = pd.DataFrame({
    'Loja 1': Quantidade1,
    'Loja 2': Quantidade2,
    'Loja 3': Quantidade3,
    'Loja 4': Quantidade4
})

df_quantidade = df_quantidade.fillna(0)

print(df_quantidade)
```

```python
import matplotlib.pyplot as plt

df_quantidade.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.title('Quantidade de Categorias de Produto por Loja')
plt.xlabel('Categoria de Produto')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
media1 = loja['Avaliação da compra'].mean()
media2 = loja2['Avaliação da compra'].mean()
media3 = loja3['Avaliação da compra'].mean()
media4 = loja4['Avaliação da compra'].mean()

print(f'Média loja 1: {media1:.2f}')
print(f'Média loja 2: {media2:.2f}')
print(f'Média loja 3: {media3:.2f}')
print(f'Média loja 4: {media4:.2f}')
```

```python
import matplotlib.pyplot as plt

medias = [media1, media2, media3, media4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.barh(lojas, medias, color='skyblue')

plt.title('Média da Avaliação da Compra por Loja')
plt.xlabel('Média da Avaliação')
plt.ylabel('Lojas')

for i, v in enumerate(medias):
    plt.text(v + 0.02, i, f'{v:.2f}', va='center', color='black')

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja


```python
frete_loja1 = loja["Frete"].mean()
frete_loja2 = loja2["Frete"].mean()
frete_loja3 = loja3["Frete"].mean()
frete_loja4 = loja4["Frete"].mean()

print(f'Média frete loja 1: {frete_loja1:.2f}')
print(f'Média frete loja 2: {frete_loja2:.2f}')
print(f'Média frete loja 3: {frete_loja3:.2f}')
print(f'Média frete loja 4: {frete_loja4:.2f}')
```

```python
import matplotlib.pyplot as plt

faturamentos = [frete_loja1, frete_loja2, frete_loja3, frete_loja4]
lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]

plt.figure(figsize=(8, 8))
plt.pie(faturamentos, labels=lojas, autopct='%1.1f%%', startangle=90, colors=plt.cm.Pastel1.colors)

plt.title("Média de frete gasto por loja", fontsize=16)

plt.show()
```

Relatório de Análise:

Após uma análise detalhada dos dados fornecidos, cheguei à conclusão de que a Loja 1 deve ser considerada para venda. Embora essa loja tenha o maior faturamento entre as quatro opções, ela não se destaca de maneira significativa em relação à Loja 2, que tem um faturamento um pouco inferior, mas apresenta uma melhor avaliação de clientes.

Faturamento vs. Satisfação do Cliente:
Apesar do alto faturamento da Loja 1, sua baixa avaliação (3.98) indica que a experiência do cliente está comprometida, o que pode afetar a satisfação e fidelização do público. Isso sugere que, embora as vendas sejam boas, a loja pode enfrentar dificuldades a longo prazo devido à insatisfação dos clientes.

Impacto do Frete:
O alto custo de frete da Loja 1 (R$ 34.69), o mais elevado entre as lojas, pode levar à recusa de clientes. Em um mercado competitivo, consumidores comparam não apenas o preço dos produtos, mas também os custos de envio. Esse fator, aliado à baixa avaliação dos clientes, sugere que a loja tem dificuldades em competir em termos de custo-benefício.

Análise da Loja 4:
Apesar do menor faturamento da Loja 4, ela se destaca com alta avaliação (4.00) e baixo custo de frete (R$ 31.28), tornando-a mais competitiva no mercado. Essa combinação de satisfação dos clientes e menor custo total a torna uma opção mais atraente, com maior potencial de crescimento a longo prazo.

Conclusão:
Em resumo a Loja 1 deve ser vendida devido ao alto custo de frete e baixa avaliação dos clientes, o que pode dificultar seu crescimento e fidelização. Em contraste, a Loja 4, apesar de ter o menor faturamento, se destaca pela alta satisfação dos clientes e baixo custo de frete, tornando-a uma opção mais estratégica para um novo empreendimento, com maior potencial de sucesso a longo prazo.
